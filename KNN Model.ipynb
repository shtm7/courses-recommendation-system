### Recommendation Systems : User-based Collaborative Filtering using Nearest Neighbors

```python
df.head()
```

```python
rate.head()
```

```python
Mean = rate.groupby(by="student_id",as_index=False)['rate_score'].mean()
Mean.head()
```

```python
Rating_avg = pd.merge(rate, Mean, on='student_id')
Rating_avg.head()
```

```python
Rating_avg['adg_rating']=Rating_avg['rate_score_x']-Rating_avg['rate_score_y']
Rating_avg.head()
```

```python
# Pivot Table
final = pd.pivot_table(Rating_avg,values='adg_rating',index='student_id',columns='course_id')
final
```

```python
# Replacing NaN by Average
final_rate = final.fillna(final.mean(axis=0))

# Replacing NaN by user Average
final_student = final.apply(lambda row: row.fillna(row.mean()), axis=1)
```

```python
final_rate
```

```python
final_student
```

```python
# user similarity on replacing NAN by student avg
b = cosine_similarity(final_student)
np.fill_diagonal(b, 0 )
similarity_with_student = pd.DataFrame(b, index=final_student.index)
similarity_with_student.columns=final_student.index
similarity_with_student.head()
```

```python
# user similarity on replacing NAN by item(movie) avg
cosine = cosine_similarity(final_rate)
np.fill_diagonal(cosine, 0 )
similarity_with_course = pd.DataFrame(cosine,index=final_rate.index)
similarity_with_course.columns=final_rate.index
similarity_with_course.head()
```

```python
def find_n_neighbours(df1,n):
    order = np.argsort(df1.values, axis=1)[:, :n]
    df1 = df1.apply(lambda x: pd.Series(x.sort_values(ascending=False)
           .iloc[:n].index, 
          index=['Top{}'.format(i) for i in range(1, n+1)]), axis=1)
    return df1
```

```python
# Top 20 neighbours for each student
sim_student_20_u = find_n_neighbours(similarity_with_student,20)
sim_student_20_u.head()
```

```python
# Top 20 neighbours for each student
sim_course_20_m = find_n_neighbours(similarity_with_course,20)
sim_course_20_m.head()
```

```python
def get_student_similar_courses( student1, student2 ):
    common_courses = Rating_avg[Rating_avg.student_id == student1].merge(
    Rating_avg[Rating_avg.student_id == student2],
    on = "course_id",
    how = "inner" )
    return common_courses.merge( df, on = 'course_id' )
```

```python
def stuent_curse_score(student,course):
    a = sim_course_20_m[sim_course_20_m.index==student].values
    b = a.squeeze().tolist()
    c = final_rate.loc[:,course]
    d = c[c.index.isin(b)]
    f = d[d.notnull()]
    avg_student = Mean.loc[Mean['student_id'] == student,'rate_score'].values[0]
    index = f.index.values.squeeze().tolist()
    corr = similarity_with_course.loc[student,index]
    fin = pd.concat([f, corr], axis=1)
    fin.columns = ['adg_score','correlation']
    fin['score']=fin.apply(lambda x:x['adg_score'] * x['correlation'],axis=1)
    nume = fin['score'].sum()
    deno = fin['correlation'].sum()
    final_score = avg_student + (nume/deno)
    return final_score
```

```python
score = stuent_curse_score(97,35)
print("The score is",score)
```

```python
Rating_avg = Rating_avg.astype({"course_id": str})
Course_student = Rating_avg.groupby(by = 'student_id')['course_id'].apply(lambda x:','.join(x))
```

```python
def Student_course_score1(student):
    course_seen_by_student = final.columns[final[final.index==student].notna().any()].tolist()
    a = sim_course_20_m[sim_course_20_m.index==student].values
    b = a.squeeze().tolist()
    d = Course_student[Course_student.index.isin(b)]
    l = ','.join(d.values)
    courses_seen_by_similar_students = l.split(',')
    courses_under_consideration = list(set(courses_seen_by_similar_students)-set(list(map(str, course_seen_by_student))))
    courses_under_consideration = list(map(int, courses_under_consideration))
    score = []
    for item in courses_under_consideration:
        c = final_rate.loc[:,item]
        d = c[c.index.isin(b)]
        f = d[d.notnull()]
        avg_student = Mean.loc[Mean['student_id'] == student,'rate_score'].values[0]
        index = f.index.values.squeeze().tolist()
        corr = similarity_with_course.loc[student,index]
        fin = pd.concat([f, corr], axis=1)
        fin.columns = ['adg_score','correlation']
        fin['score']=fin.apply(lambda x:x['adg_score'] * x['correlation'],axis=1)
        nume = fin['score'].sum()
        deno = fin['correlation'].sum()
        final_score = avg_student + (nume/deno)
        score.append(final_score)
    data = pd.DataFrame({'course_id':courses_under_consideration,'score':score})
    top_5_recommendation = data.sort_values(by='score',ascending=False).head(5)
#     print(top_5_recommendation)
#     print('----')
    Course_Name = top_5_recommendation.merge(df, how='inner', on='course_id')
    Courses_Names = Course_Name.course_name.values.tolist()
    Courses_Names = [Courses_Names[i] for i, x in enumerate(Courses_Names) if x not in Courses_Names[i+1:]]
    return Courses_Names
```

### Prediction

```python
student = int(97)
predicted_courses = Student_course_score1(student)
predicted_courses
print(" ")
print("The Recommendations for student Id : 97")
print("   ")
for i in predicted_courses:
    print(i)
```
